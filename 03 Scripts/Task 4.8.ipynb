### 5.The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.


```python
# makeing flags for max order customers
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking to make sure everything adds up
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# looking at a basic statistic 
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

### 6.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:


```python
# makeing a new column for mean prices by user id and the mean of there prices
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking to make sure it looks alright
ords_prods_merge.head(100).drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)
```

```python
# Time to derive new variables
ords_prods_merge.loc[ords_prods_merge['mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Checking to make sure everything adds up
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

### 7.In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the

```python
#Creating another column for median orders for user id by meadian days since prior order
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Checking to make sure it looks alright
ords_prods_merge.head(100).drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)
```

```python
# makeing flags for median orders
ords_prods_merge.loc[ords_prods_merge['median_order'] > 20, 'Order_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <= 20), 'Order_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10, 'Order_flag'] = 'Frequent customer'
```

```python
#Checking to make sure everything adds up
ords_prods_merge['Order_flag'].value_counts(dropna=False)
```
